Splitting the Features and Target

```python
features = heart_data.drop(columns='target', axis=1)
target = heart_data['target']
print(features)
print(target)
```

Data Standardization

```python
scaler = StandardScaler()
scaler.fit(features)
standardized_data = scaler.transform(features)
print(standardized_data)
features = standardized_data
target = heart_data['target']
print(features)
print(target)
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(features, target, test_size=0.2, stratify = target, random_state=3)
```

Comparing the performance of the models using Scratch models

```python
# list of models

classifiers = [("Logistic Regression", Log_Reg.logistic_regression(learning_rate=0.01, no_of_iterations=1000)),
              ("SVM", SVM.SVM_Classifier(learning_rate=0.001, no_of_iterations=1000, lambda_parameter=0.01)),
              ("Naive Bayes", GNB.NaiveBayesClassifier()),
              ("KNN", KNN.KNN_Classifier(k=5)),
              ("Random Forest", RF.RandomForestClassifier()),
              ("Hypersphere", Hypersphere.HypersphereClassifier()),
              ("KMeans", ScratchKMeans.KMeans_Clustering(n_clusters=2))]

def compare_models_train_test():

  for name,model in classifiers:
      if name == "KNN":
          # KNN does not have fit()
          train_pred = model.predict(X_train, Y_train, X_train)
          test_pred  = model.predict(X_train, Y_train, X_test)
      elif name == "Random Forest":
          # Convert Y_train to numpy array for Random Forest's bootstrap method
          model.fit(X_train, Y_train.to_numpy())
          train_pred = model.predict(X_train)
          test_pred  = model.predict(X_test)
      elif name == "KMeans":
          # KMeans is a clustering algorithm, not directly comparable with classification accuracy
          # We will just fit it and skip accuracy calculation for now.
          model.fit(X_train)
          print(f"{name} → Fit completed (clustering model), skipping accuracy.")
          continue # Skip accuracy calculation for KMeans
      else:
          model.fit(X_train, Y_train)
          train_pred = model.predict(X_train)
          test_pred  = model.predict(X_test)

      train_acc = accuracy_score(Y_train, train_pred)
      test_acc  = accuracy_score(Y_test, test_pred)

      print(f"{name} → Train Accuracy: {train_acc:.4f}, Test Accuracy: {test_acc:.4f}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

def compare_binary_metrices_of_sklearn_models():
    print("\n--- Super Simple Scikit-learn Train/Test Evaluation ---")

    for model in models:
        model_name = type(model).__name__
        print(f"\nEvaluating {model_name}...")

        if isinstance(model, KMeans): # KMeans is a clustering algorithm
            print(f"{model_name} is a clustering model. Skipping classification metrics.")
            continue

        # 1. Train the model
        model.fit(X_train, Y_train)

        # 2. Get predictions for training and testing data
        train_pred = model.predict(X_train)
        test_pred = model.predict(X_test)

        # 3. Calculate and print simple metrics for training data
        print(f"  --- Training Metrics for {model_name} ---")
        print(f"    Accuracy: {accuracy_score(Y_train, train_pred):.4f}")
        print(f"    Precision: {precision_score(Y_train, train_pred, zero_division=0):.4f}")
        print(f"    Recall: {recall_score(Y_train, train_pred, zero_division=0):.4f}")
        print(f"    F1-Score: {f1_score(Y_train, train_pred, zero_division=0):.4f}")

        # 4. Calculate and print simple metrics for testing data
        print(f"  --- Testing Metrics for {model_name} ---")
        print(f"    Accuracy: {accuracy_score(Y_test, test_pred):.4f}")
        print(f"    Precision: {precision_score(Y_test, test_pred, zero_division=0):.4f}")
        print(f"    Recall: {recall_score(Y_test, test_pred, zero_division=0):.4f}")
        print(f"    F1-Score: {f1_score(Y_test, test_pred, zero_division=0):.4f}")
```

Cross Validation (SKlearn models)

```python
# list of models (excluding KMeans for cross-validation as it's a clustering algorithm)
models_cv = [LogisticRegression(max_iter=1000), SVC(kernel='linear'), GaussianNB(), KNeighborsClassifier(), RandomForestClassifier() ]
```

```python
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, make_scorer

def compare_models_cross_validation():
    # Define custom scorers with zero_division handled
    scoring = {
        'accuracy': make_scorer(accuracy_score),
        'precision': make_scorer(precision_score, zero_division=0),
        'recall': make_scorer(recall_score, zero_division=0),
        'f1_score': make_scorer(f1_score, zero_division=0)
    }

    print("\n--- Detailed Cross-Validation Metrics (SKlearn models) ---")

    for model in models_cv:
        model_name = type(model).__name__
        print(f"\nEvaluating {model_name}...")

        # Use StratifiedKFold for classification tasks to maintain class distribution
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

        # Perform cross-validation and get scores for multiple metrics
        # Use the custom scorers from the 'scoring' dictionary
        cv_results = cross_val_score(model, features, target, cv=cv, scoring=scoring['accuracy'])
        cv_precision = cross_val_score(model, features, target, cv=cv, scoring=scoring['precision'])
        cv_recall = cross_val_score(model, features, target, cv=cv, scoring=scoring['recall'])
        cv_f1 = cross_val_score(model, features, target, cv=cv, scoring=scoring['f1_score'])

        print(f"  Accuracy per fold: {cv_results}")
        print(f"  Precision per fold: {cv_precision}")
        print(f"  Recall per fold: {cv_recall}")
        print(f"  F1-Score per fold: {cv_f1}")

        print(f"  --- Aggregated Metrics for {model_name} ---")
        print(f"    Mean Accuracy: {cv_results.mean():.4f} (Std: {cv_results.std():.4f})")
        print(f"    Mean Precision: {cv_precision.mean():.4f} (Std: {cv_precision.std():.4f})")
        print(f"    Mean Recall: {cv_recall.mean():.4f} (Std: {cv_recall.std():.4f})")
        print(f"    Mean F1-Score: {cv_f1.mean():.4f} (Std: {cv_f1.std():.4f})")
        print("----------------------------------------------")
```

```python
compare_models_cross_validation()
```

Hyperparameter Tuning

GridSearchCV

```python
from sklearn.model_selection import GridSearchCV

def simplified_hyperparameter_tuning(X_train, Y_train):
    # Define the models and their respective parameter grids
    tuning_models = {
        'Logistic Regression': {
            'model': LogisticRegression(max_iter=1000),
            'params': {'C': [0.1, 1, 10, 100]}
        },
        'SVC': {
            'model': SVC(),
            'params': {'kernel': ['linear', 'rbf'], 'C': [0.1, 1, 10]}
        },
        'GaussianNB': {
            'model': GaussianNB(),
            'params': {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}
        },
        'KNeighborsClassifier': {
            'model': KNeighborsClassifier(),
            'params': {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance']}
        },
        'RandomForestClassifier': {
            'model': RandomForestClassifier(random_state=42),
            'params': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20]}
        }
    }

    print("Starting simplified hyperparameter tuning...")
    results = []

    for name, config in tuning_models.items():
        print(f"\nTuning {name}...")
        grid_search = GridSearchCV(config['model'], config['params'], cv=5, scoring='accuracy', n_jobs=-1)
        grid_search.fit(X_train, Y_train)

        best_params = grid_search.best_params_
        best_score = grid_search.best_score_

        print(f"  Best parameters for {name}: {best_params}")
        print(f"  Best cross-validation accuracy for {name}: {best_score:.4f}")

        results.append({
            'model': name,
            'best_params': best_params,
            'best_score': best_score
        })

    print("\nHyperparameter tuning complete!")
    return results

# Run the tuning function
tuning_results = simplified_hyperparameter_tuning(X_train, Y_train)

# Display results as a DataFrame for easy viewing
import pandas as pd
display(pd.DataFrame(tuning_results))
```

RandomizedSearchCV

```python
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import uniform, randint

def simplified_randomized_search_tuning(X_train, Y_train):
    # Define the models and their respective parameter distributions for RandomizedSearchCV
    tuning_models = {
        'Logistic Regression': {
            'model': LogisticRegression(max_iter=1000),
            'param_distributions': {'C': [0.1, 1, 10, 100]}
        },
        'SVC': {
            'model': SVC(),
            'param_distributions': {'kernel': ['linear', 'rbf'], 'C': [0.1, 1, 10]}
        },
        'GaussianNB': {
            'model': GaussianNB(),
            'param_distributions': {'var_smoothing': [1e-9, 1e-8, 1e-7, 1e-6, 1e-5]}
        },
        'KNeighborsClassifier': {
            'model': KNeighborsClassifier(),
            'param_distributions': {'n_neighbors': [3, 5, 7, 9], 'weights': ['uniform', 'distance']}
        },
        'RandomForestClassifier': {
            'model': RandomForestClassifier(random_state=42),
            'param_distributions': {'n_estimators': [50, 100, 200], 'max_depth': [None, 10, 20], 'min_samples_split': [2, 5, 10]}
        }
    }

    print("Starting simplified randomized search hyperparameter tuning...")
    results = []

    for name, config in tuning_models.items():
        print(f"\nTuning {name}...")
        # Use n_iter=10 for a reasonable number of iterations, cv=5 for cross-validation
        random_search = RandomizedSearchCV(config['model'], config['param_distributions'], n_iter=10, cv=5, scoring='accuracy', n_jobs=-1, random_state=42)
        random_search.fit(X_train, Y_train)

        best_params = random_search.best_params_
        best_score = random_search.best_score_

        print(f"  Best parameters for {name}: {best_params}")
        print(f"  Best cross-validation accuracy for {name}: {best_score:.4f}")

        results.append({
            'model': name,
            'best_params': best_params,
            'best_score': best_score
        })

    print("\nRandomized search hyperparameter tuning complete!")
    return results

# Run the tuning function
random_tuning_results = simplified_randomized_search_tuning(X_train, Y_train)

# Display results as a DataFrame for easy viewing
import pandas as pd
display(pd.DataFrame(random_tuning_results))
```

```python
import pandas as pd
import numpy as np

def get_cv_metrics_for_plotting():
    results = []
    for model in models_cv:
        model_name = type(model).__name__

        # Use StratifiedKFold for classification tasks to maintain class distribution
        cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

        # Define custom scorers with zero_division handled
        scoring = {
            'accuracy': make_scorer(accuracy_score),
            'precision': make_scorer(precision_score, zero_division=0),
            'recall': make_scorer(recall_score, zero_division=0),
            'f1_score': make_scorer(f1_score, zero_division=0)
        }

        # Perform cross-validation and get scores for multiple metrics
        cv_accuracy = cross_val_score(model, features, target, cv=cv, scoring=scoring['accuracy'])
        cv_precision = cross_val_score(model, features, target, cv=cv, scoring=scoring['precision'])
        cv_recall = cross_val_score(model, features, target, cv=cv, scoring=scoring['recall'])
        cv_f1 = cross_val_score(model, features, target, cv=cv, scoring=scoring['f1_score'])

        results.append({
            'Model': model_name,
            'Mean Accuracy': np.mean(cv_accuracy),
            'Mean Precision': np.mean(cv_precision),
            'Mean Recall': np.mean(cv_recall),
            'Mean F1-Score': np.mean(cv_f1)
        })
    return pd.DataFrame(results)

cv_metrics_df = get_cv_metrics_for_plotting()
print(cv_metrics_df)
```

```python
import matplotlib.pyplot as plt

# Prepare data for plotting
metrics = ['Mean Accuracy', 'Mean Precision', 'Mean Recall', 'Mean F1-Score']
models = cv_metrics_df['Model']

x = np.arange(len(models))  # the label locations
width = 0.2  # the width of the bars

plt.figure(figsize=(15, 8))

for i, metric in enumerate(metrics):
    plt.bar(x + i * width, cv_metrics_df[metric], width, label=metric)

plt.xlabel('Models')
plt.ylabel('Score')
plt.title('Comparison of Mean Cross-Validation Metrics for SKlearn Models')
plt.xticks(x + width * (len(metrics) - 1) / 2, models, rotation=45, ha='right')
plt.legend(loc='lower right')
plt.ylim(0.7, 1.0) # Set a reasonable y-limit to better visualize differences
plt.tight_layout()
plt.show()
```
